```python
import docstring_parser
import inspect

from dataclasses import dataclass
from attrs import define


@dataclass
class A:
    "dsadsasa"

    x: list[int]
    y: int


@define
class B:
    "dsadsasa"

    x: list[int]
    y: int


a = A([1, 2, 3], 890)
b = B([1, 2, 3], 890)
```

```python
docstring = docstring_parser.parse(
    inspect.getdoc(a), docstring_parser.common.DocstringStyle.GOOGLE
)
print(docstring.short_description)
print(docstring.long_description)
```

```python
docstring = docstring_parser.parse(
    inspect.getdoc(b), docstring_parser.common.DocstringStyle.GOOGLE
)
print(docstring.short_description)
print(docstring.long_description)
```
