```python
normal = "normal = "
reordered = "reordered = "
sort_column = "score2"
top_k = 100


def get_delta_results(filepath):
    with pathlib.Path(TMP_DIR / filepath).open("rb") as f:
        results = pickle.load(f)
        results = list(results.values())[0]
        return results


def compare(results, sort_column):
    if isinstance(results, str):
        print(results)
        results = get_delta_results(results)
    results = results.sort_values([sort_column], ascending=False)
    results = results[results["count"] > 0]
    seq = np.arange(1, len(results) + 1)
    tmp = results["column"].str.contains("reordered_")
    print(
        f"mean column rank\n"
        f"{normal:<15}{seq[~tmp].mean()} --- {sum(seq[~tmp])} \n"
        f"{reordered:<15}{seq[tmp].mean()} --- {sum(seq[tmp])}\n"
    )


def compare2(results, sort_column, top_k=None):
    if isinstance(results, str):
        print(results)
        results = get_delta_results(results)
    results = results.sort_values([sort_column], ascending=False)
    ranks = pd.DataFrame(
        {
            "column": results["column"],
            "rank": np.arange(len(results)) + 1,
            "count": results["count"],
            "is_reordered": results["column"].str.contains("reordered_"),
        }
    )
    ranks.reset_index(inplace=True, drop=True)
    ranks.loc[ranks["count"].values == 0, "rank"] = len(ranks)
    print(
        f"mean column rank\n"
        f'{normal:<15}{ranks[~ranks["is_reordered"]]["rank"].mean()}'
        f' --- {len(ranks[~ranks["is_reordered"]])} \n'
        f'{reordered:<15}{ranks[ranks["is_reordered"]]["rank"].mean()}'
        f' --- {len(ranks[ranks["is_reordered"]])}'
    )
    if top_k is not None:
        print(
            f'top_k({top_k}) {normal:<15}{ranks[~ranks["is_reordered"]].iloc[:top_k]["rank"].mean()}'
            f' --- {len(ranks[~ranks["is_reordered"]].iloc[:top_k])} \n'
            f'top_k({top_k}) {reordered:<15}{ranks[ranks["is_reordered"]].iloc[:top_k]["rank"].mean()}'
            f' --- {len(ranks[ranks["is_reordered"]].iloc[:top_k])} \n'
        )


compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_3_eval_scope_local.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_3_eval_scope_global.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_None_eval_scope_local.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_None_eval_scope_global.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_3_eval_scope_local.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_3_eval_scope_global.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_None_eval_scope_local.pkl",
    sort_column,
)
compare(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_None_eval_scope_global.pkl",
    sort_column,
)

print("##########")
print("##########")
print("##########")
print("##########")

compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_3_eval_scope_local.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_3_eval_scope_global.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_None_eval_scope_local.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_100_allowed_randomness_0.1_max_n_attrs_None_eval_scope_global.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_3_eval_scope_local.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_3_eval_scope_global.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_None_eval_scope_local.pkl",
    sort_column,
    top_k=top_k,
)
compare2(
    "compare/bireducts_daab_reordered_cols_n_1000_candidate_attrs_30_allowed_randomness_0.2_max_n_attrs_None_eval_scope_global.pkl",
    sort_column,
    top_k=top_k,
)
```

```python
DATA_DIR = pathlib.Path(config.DATA_DIR)
FILEPATH_IN = DATA_DIR / "train_utf_with_reordered_cols.csv"
SEP = ";"


df = pd.read_csv(FILEPATH_IN, sep=SEP)
df_dec = df.pop("target")
df = df.astype("category")
df = df.apply(lambda x: x.cat.codes)
df_dec_bin = df_dec == "interested"
```

```python
n_estimators = 1000
learning_rate = 0.01
max_depth = 3
cl = xgb.XGBClassifier(
    eval_metric="logloss",
    use_label_encoder=False,
    n_estimators=n_estimators,
    learning_rate=learning_rate,
    max_depth=max_depth,
)
cl.fit(df, df_dec_bin)
```

```python
xresults = pd.DataFrame(
    {
        "column": df.columns,
        "score": cl.feature_importances_,
        "count": cl.feature_importances_ > 0,
    }
)
compare2(xresults, "score", top_k=100)
```

```python
q = xresults[cl.feature_importances_ > 0]
q.sort_values("score", ascending=False)
```
