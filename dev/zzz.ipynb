```python
import numba
import numpy as np
import numpy.typing as npt


@numba.njit
def _get_distribution(
    groups: npt.NDArray[np.int64],
    groups_count: int,
    values: npt.NDArray[np.int64],
    values_count: int,
) -> npt.NDArray[np.int64]:
    """
    Compute decision distribution within groups of objects
    """
    result = np.zeros(shape=(groups_count, values_count), dtype=np.int64)
    nrow = groups.shape[0]
    for i in range(nrow):
        result[groups[i], values[i]] += 1
    return result
```

```python
groups = np.array([1, 2, 2, 3, 1, 1, 1, 2, 3])
values = np.array([0, 1, 1, 1, 1, 1, 0, 0, 1])
_get_distribution(groups, 7, values, 2)
```

```python
groups_count_1 = 100000
values_count_1 = 5
groups_1 = np.random.randint(0, groups_count_1, size=10000000)
values_1 = np.random.randint(0, values_count_1, size=10000000)
```

```python
%%timeit
_get_distribution(groups_1, groups_count_1, values_1, values_count_1)
```

```python
@numba.njit
def _get_distribution_v2(
    groups: npt.NDArray[np.int64],
    values: npt.NDArray[np.int64],
    result: npt.NDArray[np.int64],
) -> None:
    result.fill(0)
    nrow = groups.shape[0]
    for i in range(nrow):
        result[groups[i], values[i]] += 1
```

```python
result_test = np.zeros(shape=(10, 2), dtype=np.int64)
_get_distribution_v2(groups, values, result_test)
result_test
```
